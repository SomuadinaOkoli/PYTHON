```python
#        Calculator application
from tkinter import *

class Calculator:
    # Building the GUI
    def __init__(self):
        self.GUI = Tk()
        self.GUI.title("Calculator")
        self.GUI.geometry("300x300")
        
        # Setting the display widget
        self.display_frame = Frame(self.GUI, bg = "white", height= 100, width = 300)
        self.display_frame.place(x = 0, y=0)

        # Creating a frame for the number buttons
        self.num_frame = Frame(self.GUI,bg= "black", height = 200, width = 200)
        self.num_frame.place( x = 0, y = 100)
        
        # Creating the number buttons
        self.but_1 = Button(self.num_frame, text = "1", bg = "grey", width = 5, command = lambda: self.insertToEntry(1))
        self.but_1.place(x = 20, y = 15)

        self.but_2 = Button(self.num_frame, text = "2", bg = "grey", width = 5, command = lambda: self.insertToEntry(2))
        self.but_2.place(x = 80, y = 15)

        self.but_3 = Button(self.num_frame, text = "3", bg = "grey", width = 5, command = lambda: self.insertToEntry(3))
        self.but_3.place(x = 140, y = 15)

        self.but_4 = Button(self.num_frame, text = "4", bg = "grey", width = 5, command = lambda: self.insertToEntry(4))
        self.but_4.place(x = 20, y = 55)

        self.but_5 = Button(self.num_frame, text = "5", bg = "grey", width = 5, command = lambda: self.insertToEntry(5))
        self.but_5.place(x = 80, y = 55)

        self.but_6 = Button(self.num_frame, text = "6", bg = "grey", width = 5, command = lambda: self.insertToEntry(6))
        self.but_6.place(x = 140, y = 55)

        self.but_7 = Button(self.num_frame, text = "7", bg = "grey", width = 5, command = lambda: self.insertToEntry(7))
        self.but_7.place(x = 20, y = 95)

        self.but_8 = Button(self.num_frame, text = "8", bg = "grey", width = 5, command = lambda: self.insertToEntry(8))
        self.but_8.place(x = 80, y = 95)

        self.but_9 = Button(self.num_frame, text = "9", bg = "grey", width = 5, command = lambda: self.insertToEntry(9))
        self.but_9.place(x = 140, y = 95)

        self.but_decimal = Button(self.num_frame, text = ".", bg = "grey", width = 5, command = lambda: self.insertToEntry("."))
        self.but_decimal.place(x = 20, y = 135)

        self.but_zero = Button(self.num_frame, text = "0", bg = "grey", width = 5, command = lambda: self.insertToEntry(0))
        self.but_zero.place(x = 80, y = 135)

        self.but_right = Button(self.num_frame, text = "Right", bg = "grey", width = 5, command = self.cursorRight)
        self.but_right.place(x = 140, y = 135)

        self.but_del = Button(self.num_frame, text = "Del", bg = "grey", width = 5, command = self.delete)
        self.but_del.place(x = 20, y = 170)

        self.but_left = Button(self.num_frame, text = "Left", bg = "grey", width = 5, command = self.cursorLeft)
        self.but_left.place(x = 140, y = 170)
    

        # Creating the frame for the mathematical operators
        self.operator_frame = Frame(self.GUI, bg = "grey", height = 200, width = 100)
        self.operator_frame.place(x = 200, y = 100)
        # Creating the button for the operators
        self.but_add = Button(self.operator_frame, text = '+', width = 3, command = self.buttonAdd)
        self.but_add.place(x = 20, y = 15)

        self.but_subtract = Button(self.operator_frame, text = '-', width = 3, command = self.buttonSubtract)
        self.but_subtract.place(x = 60, y = 15)

        self.but_multiply = Button(self.operator_frame, text = 'x', width = 3, command = self.buttonMultiply)
        self.but_multiply.place(x = 20, y = 55)

        self.but_divide = Button(self.operator_frame, text = '/', width = 3, command = self.buttonDivision)
        self.but_divide.place(x = 60, y = 55)

        self.but_modulos = Button(self.operator_frame, text = '%', width = 3, command = self.buttonModulos)
        self.but_modulos.place(x = 20, y = 95)

        self.but_clear = Button(self.operator_frame, text = 'C', width = 3, command = self.buttonClear)
        self.but_clear.place(x = 60, y = 95)

        # Creating the frame that will contain the entry fields for num1 and num2
        self.entry_frame = Frame(self.GUI, height = 50, width = 300)
        self.entry_frame.place(x = 0, y = 50)
        # Creating the entry fields for num1 and num2
        self.entry_num1 = Entry(self.entry_frame)
        self.entry_num1.place(x = 5, y = 25)

        self.entry_num2 = Entry(self.entry_frame)
        self.entry_num2.place(x = 170, y = 25)

        # Creating a variable that will be used to determine which entry field should receive values from the number buttons
        self.cursor = self.entry_num1 # What this means is that, by default, entry_num1 will be the first entry field to receive values from the buttons
        
        # Creating the top display label
        self.display = Label(self.GUI, bg = "white", height = 1, font = ("Calibre", 20, "bold"))
        self.display.place(x = 125, y = 13)


        
        self.GUI.mainloop()

    
    # Methods of the Calculator class

    # Mathematical operator methods
    def buttonAdd(self): # This button method will carry out the summation of the two entry field
        num1 = float(self.entry_num1.get())
        num2 = float(self.entry_num2.get())
        self.display.config(text = f"{num1 + num2}") # This will configure the display label's text to the result
        
    def buttonSubtract(self):# This button method will carry out the subtraction of the two entry field
        num1 = float(self.entry_num1.get())
        num2 = float(self.entry_num2.get())
        self.display.config(text = f"{num1 - num2}") # This will configure the display label's text to the result
        
    def buttonMultiply(self):# This button method will carry out the multiplication of the two entry field
        num1 = float(self.entry_num1.get())
        num2 = float(self.entry_num2.get())
        self.display.config(text = f"{num1 * num2}") # This will configure the display label's text to the result
        
    def buttonDivision(self): # This button method will carry out the division of the two entry field
        num1 = float(self.entry_num1.get())
        num2 = float(self.entry_num2.get())
        self.display.config(text = f"{num1 / num2}") # This will configure the display label's text to the result

    def buttonModulos(self): # This button method will display the modulo of the two entry field
        num1 = float(self.entry_num1.get())
        num2 = float(self.entry_num2.get())
        self.display.config(text = f"{int(num1 % num2)}") # This will configure the display label's text to the result

    def buttonClear(self): # This button method will delete all the numbers in the two entry fields
        self.entry_num1.delete(0, len(self.entry_num1.get())) # For entry_num1
        self.entry_num2.delete(0, len(self.entry_num2.get())) # For entry_num2
    
    # Numbers and entries methods
    def insertToEntry(self, text):# This function will be used to insert the number gotten from a button into the current entry field
        self.cursor.insert(len(self.cursor.get()), text)
        
    def delete(self):# This function would delete the last item in an entry field
        #popLast = self.cursor.get()[:len(self.cursor.get()) - 1]
        
        self.cursor.delete(len(self.cursor.get()) - 1, len(self.cursor.get()))
    
    def cursorRight(self):# This function will move the cursor to the right entry field
        self.cursor = self.entry_num2
    def cursorLeft(self):# This function will move the cursor to the left entry field
        self.cursor = self.entry_num1
    
cObj = Calculator()                
```
